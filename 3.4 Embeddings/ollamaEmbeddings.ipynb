```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="llamma2")
embeddings
```

```python
text_embed = embeddings.embed_documents(
    [
        "Alpha is first letter in Greek alphabet",
        "Beta is second letter in Greek alphabet"

    ]
)
text_embed
```
